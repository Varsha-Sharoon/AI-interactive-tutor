**This project converts handwritten** **or** **typed notes** **into simple**,**friendly conversational explanations using Groq‚Äôs LLaMA 3 model**.

OBJECTIVE :       

Students often struggle to understand complex notes or textbook   explanations. This project introduces an AI-based learning assistant that simplifies learning by converting uploaded text or images of notes into clear, conversational explanations. Using Large Language Models (LLMs) and Optical Character Recognition (OCR), the system explains the topic in simple English, interacts conversationally with the student, and asks short questions to test understanding. If the student cannot answer, the AI automatically re-explains the concept with simpler examples. This creates a personalized and adaptive self-learning experience for students.

# Step  1 :

*Installing required libraries ;*

```python
# Installing  libraries
!pip install requests pytesseract pillow groq
# Import libraries
import requests
from PIL import Image
import pytesseract
from groq import Groq
from google.colab import files
import getpass
```

```python
# Configure Groq API key
import getpass
from groq import Groq

GROQ_API_KEY = getpass.getpass("Enter your Groq API key: ")
client = Groq(api_key=GROQ_API_KEY)
```

# STEP 2 :

**Uploading Text or Image & Extracting Text**

```python
print('Please upload your notes file(can be .txt,.pdf,.jpg, or .png)')
uploaded = files.upload()  # Choose a file when prompted
for filename in uploaded.keys():
    print(f"\nüìò Uploaded File: {filename}")

    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        image = Image.open(filename)
        text = pytesseract.image_to_string(image)
    elif filename.lower().endswith('.txt'):
        with open(filename, 'r', encoding='utf-8') as f:
            text = f.read()
    else:
        text = "Unsupported file type. Please upload an image or text file."
print("\nüìù Extracted Text:\n")
print(text[:2000])
```

# STEP 3 :

***SIMPLIFIED*** **AI** ***EXPLANATION***

```python
prompt = f"Explain the following notes in simple, friendly conversational English as if teaching a beginner:\n\n{text}"
response = client.chat.completions.create(
model="llama-3.1-8b-instant", # Using a Groq model
messages=[{"role": "user", "content": prompt}],
max_tokens=500
)
ai_explanation = response.choices[0].message.content
print("üí¨ Simplified Explanation:\n")
print(ai_explanation)
```

# STEP 4:

**GENERATING QUIZ**

```python
# Step 4: Generate quiz questions
quiz_prompt = f"""
You are a teacher creating an interactive quiz for students based on these notes:
{text}

Generate 5 questions (mix of multiple-choice and short answers) ‚Äî but do NOT include the answers.
Format like this:

Q1. ...
A. ...
B. ...
C. ...
D. ...

Q2. ...
"""
quiz_response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": quiz_prompt}],
    max_tokens=600
)
quiz = quiz_response.choices[0].message.content
print("\nüìù Quiz Time:\n")
print(quiz)
```

```python
# Step 5: Ask user for their answers
student_answers= input("\nType your answers here (just a short summary is fine): ")
# Ask AI to evaluate the student's answers
feedback_prompt = f"""
You are a strict but friendly teacher.
Compare the student's answers with the correct meaning of the topic below:
{text}

For each question:
- Mark as Correct, Wrong, or Partially Correct.
- If wrong, give a very simple 2-line re-explanation (like you're teaching a beginner).
- Do NOT overanalyze ‚Äî just check factual correctness.

Student answers:
{student_answers}
"""
feedback_response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": feedback_prompt}],
    max_tokens=600
)
feedback = feedback_response.choices[0].message.content
print("\nüßæ Feedback / Re-Explanation:\n")
print(feedback)
```

```python
# Step 6: Re-test only for wrong answers
retest_prompt = f"""
Based on the feedback above, create 3 new quiz questions
only from the parts the student got wrong or partially correct.

Keep them simple and conversational ‚Äî do NOT give answers.
"""
retest_response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": retest_prompt}],
    max_tokens=500
)

retest_quiz = retest_response.choices[0].message.content
print("\nüîÅ Re-Test Quiz (based on wrong answers):\n")
print(retest_quiz)
```

```python
# Step 7: Option to re-explain if student can't answer
choice = input("‚ùì Can't answer any question? Type 're' to re-explain the topic, or press Enter to continue: ")

if choice.lower() == 're':
    re_explain_prompt = f"""
    Please re-explain the topic below in simpler, more conversational English.
    Make it easier for a beginner to understand, like telling a story.
    Notes:
    {text}
    """

    re_explain_response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": re_explain_prompt}],
        max_tokens=500
    )

    print("\nüîÅ Re-Explanation:\n")
    print(re_explain_response.choices[0].message.content)
```
